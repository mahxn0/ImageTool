```python
import os
import cv2
import xml.etree.ElementTree as ET
import numpy as np
```

```python
InputTextPath='/media/mahxn0/DATA/dataset/helmet_hand_data/coco.txt'
with open(InputTextPath,'r') as File_1:
    Lines=File_1.readlines()
    for Line in Lines:
        Line=Line.strip('\n')
        Filename=Line.split('/')[-1].split('.txt')[0]
        print(Line)
        ImgNames=Line[:-3]+'jpg'
        XmlNames=Line[:-3]+'xml'
        Img=cv2.imread(ImgNames)
        H,W=Img.shape[:2]
        print(H,W)
        with open(XmlNames,'w') as xml_file:
            xml_file.write('<annotation verified="no">\n')
            xml_file.write('<folder>helmet_hand_data</folder>\n')
            xml_file.write('<filename>' + str(Filename)  + '</filename>\n')
            xml_file.write('<path>' + str(ImgNames)  + '</path>\n')
            xml_file.write('<source>\n')
            xml_file.write('<database>'+'unknown'+'</database>\n')
            xml_file.write('</source>\n')
            xml_file.write('<size>\n')
            xml_file.write('<width>' + str(W) + '</width>\n')
            xml_file.write('<height>' + str(H) + '</height>\n')
            xml_file.write('<depth>3</depth>\n')
            xml_file.write('</size>\n')
            xml_file.write('<segmented>'+str(0)+'</segmented>\n')
            with open(Line,'r') as File_2:
                Txts=[Txt_a.rstrip('\n') for Txt_a in File_2.readlines()]
                for Txt in Txts:
                    Txt=Txt.split(' ')
                    width=float(Txt[3])*W
                    height=float(Txt[4])*H
                    xmin=int(float(Txt[1])*W-width/2)
                    ymin=int(float(Txt[2])*H-height/2)
                    xmax=int(xmin+width)
                    ymax=int(ymin+height)
                    print(xmin,xmax,ymin,ymax)
                    name='hand'
                    xml_file.write('<object>\n')
                    xml_file.write('<name>' + str(name) + '</name>\n')
                    xml_file.write('<pose>Unspecified</pose>\n')
                    xml_file.write('<truncated>0</truncated>\n')
                    xml_file.write('<Difficult>0</Difficult>\n')
                    xml_file.write('<bndbox>\n')
                    xml_file.write('<xmin>' + str(xmin) + '</xmin>\n')
                    xml_file.write('<ymin>' + str(ymin) + '</ymin>\n')
                    xml_file.write('<xmax>' + str(xmax) + '</xmax>\n')
                    xml_file.write('<ymax>' + str(ymax) + '</ymax>\n')
                    xml_file.write('</bndbox>\n')
                    xml_file.write('</object>\n')
            xml_file.write('</annotation>')
            File_2.close()
```
